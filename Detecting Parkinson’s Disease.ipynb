# Removing highly correlated features

```python
# Calculate the correlation matrix and take the absolute value
corr_df = data.corr().abs()

# Create a True/False mask and apply it
mask = np.triu(np.ones_like(corr_df, dtype=bool))
tri_df = corr_df.mask(mask)

# List column names of highly correlated features (r > 0.90)
to_drop = [c for c in tri_df.columns if any(tri_df[c] >  0.90)]

# Drop the features in the to_drop list
reduced_df = data.drop(to_drop, axis='columns')

# Create the correlation matrix
reduced_corr = reduced_df.corr().abs()

# Generate a mask for the upper triangle 
mask = np.triu(np.ones_like(reduced_corr, dtype=bool))

# Add the mask to the heatmap
plt.figure(figsize=(15, 15))
sns.heatmap(reduced_corr, mask=np.triu(np.ones_like(reduced_corr, dtype=bool))
, center=0, linewidths=1, annot=True, fmt=".2f")
plt.show()
```

# Feature Selection

```python
X=data.loc[:,data.columns!='status'].values
y=data.loc[:,'status'].values

```

```python
print(y[y==0].shape[0])
```

```python
print(y[y==1].shape[0])
```

# Scaling the features between -1 and 1

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
scaler=MinMaxScaler((-1,1))
X=scaler.fit_transform(X)
X
```

# Splitting the Data

```python
X_train,X_test,y_train,y_test=train_test_split(X, y,test_size=0.25,random_state=21)
```

# Model Training 

```python
from xgboost import XGBClassifier
```

```python
model=XGBClassifier()
model.fit(X_train,y_train)
```

# Prediction

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred=model.predict(X_test)
print(accuracy_score(y_test, y_pred)*100)
```
